```python
class Student:
  def __init__(self,name,age):
    self.name=name
    self.age=age

  def display(self):
    print(self.name,self.age)
s=Student("navya",20)
s.display()
```

```python
class Student:
  def __init__(self,name,age,marks):
    self.age=age
    self.name=name
    self.marks=marks

  def record(self):
    print(self.age,self.name,self.marks)

s=Student("navya",20,90)
s.record()
```

#line plot-show trends over a time

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
sales=[10,20,15,25,30]
plt.plot(days,sales)
plt.title("Sales trend")
plt.xlabel("Days")
plt.ylabel("sales")
plt.show()
```

#Bar plot-campare categories

```python
products=["A","B","C"]
sales=[100,200,150]
plt.bar(products,sales)
plt.title("Sales Comparison")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

#Pie Chart-show percentages

```python
labels=["Food","Rent","Travel"]
expense=[40,35,25]
plt.pie(expense,labels=labels,autopct="%1.1f%%")
plt.show()
```
